def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
def calculate_total_price (products):
    inventory = {}
    for product in products:
        valid_price = False
        while not valid_price:
            try:
                price = int(input(f"Enter the price of {product}s available: "))
                if price < 0:
                    raise ValueError("Invalid price! Please re-enter a non-negative value for the price.")
                valid_price = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = price
    return inventory
````

````python
def get_customer_orders(inventory):
    customer_orders = {}
    
    while True:
        try:
            num_orders = int(input("Enter the number of orders: "))
            if num_orders < 0:
                raise ValueError("The number of orders cannot be negative. Please enter a valid number.")
        except ValueError as error:
            print(f"Error: {error}")

    for _ in range(num_orders):
        while True:
            try:
                product_name = input("Enter the product name: ").strip()
                
                if product_name not in inventory:
                    raise ValueError(f"'{product_name}' is not available in the inventory. Please enter a valid product name.")
                
                if inventory[product_name] <= 0:
                    raise ValueError(f"'{product_name}' is out of stock. Please select another product.")
                
                while True:
                    try:
                        quantity = int(input(f"Enter the quantity for {product_name}: "))
                        if quantity <= 0:
                            raise ValueError("Quantity must be a positive number.")
                        if quantity > inventory[product_name]:
                            raise ValueError(f"Only {inventory[product_name]} {product_name}(s) available.")
                        break  # If valid, exit the loop
                    except ValueError as error:
                        print(f"Error: {error}")
                
                customer_orders[product_name] = quantity
                inventory[product_name] -= quantity
            except ValueError as error:
                print(f"Error: {error}")
    
    return customer_orders
````
